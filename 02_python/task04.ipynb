```python
try:
    input_a = int(input("一つ目の数値を入力してください:"))
    input_b = int(input("二つ目の数値を入力してください:"))

    # 計算処理
    result = input_a / input_b

    # 表示処理
    print(f"計算結果: {result}")
except ValueError:
    print("エラー: 数値を入力してください。")
except ZeroDivisionError:
    print("エラー: 0で割ることはできません。")
```
