```python
# 税込価格の入力
input_price = int(input("税込価格を入力してください: "))
ship_cost = 0
shipping_msg = "無料"

# 税抜価格の計算
taxprice = input_price * 1.1

# 送料の計算
match taxprice:
    case taxprice if taxprice >= 2000:
        pass
    case _:
        ship_cost = 350
        shipping_msg = ship_cost


print(f"送料は{shipping_msg}です。")
print(f"送料込税込価格は {int(taxprice + ship_cost)}円です。")
```
